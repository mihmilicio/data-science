## **PRÉ-PROCESSAMENTO**

```python
df.head(5)
```

### Selecionar as colunas que vamos utilizar

```python
df2 = df[['EdLevel', 'Country', 'Currency', 'YearsCodeNumber', 'YearsCodeProNumber', 'YearlyWageInDollar']].copy()
```

```python
df2.head()
```

```python
df3 = df2.copy()
```

### Limpando EdLevel

```python
df3.shape
```

```python
df3 = df3[(df3['EdLevel'].notna()) & (df3['EdLevel'] != 'other')]
```

```python
df3.shape
```

### Filtrando principais países

Aqui filtramos pelos 12 países mais frequentes, para evitar termos que lidar com grandes volumes de colunas na conversão para numéricos

```python
# get top 12 most frequent names
n = 12
array_countries = df3['Country'].value_counts()[:n].index.tolist()
array_countries
```

```python
df3.shape
```

```python
df3 = df3.loc[df['Country'].isin(array_countries)]
```

```python
df3.shape
```

### Resetando index

```python
df3
```

```python
df3.reset_index(inplace=True, drop=True)
df3
```

### Transformar variáveis categóricas em números

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
encoder = OneHotEncoder()
oneHotEncoder = encoder.fit_transform(df3[['EdLevel', 'Country', 'Currency']])
```

```python
oneHotEncoder.todense()
```

```python
oneHotEncoder.shape
```

```python
nome_colunas = encoder.get_feature_names(['EdLevel', 'Country', 'Currency'])
df4_encoder = pd.DataFrame(oneHotEncoder.todense(), columns= nome_colunas)
df4_encoder
```

```python
df4 = pd.concat([df3, df4_encoder], axis=1).drop(['EdLevel', 'Country', 'Currency'], axis=1)
df4.info()
```

### Reordenando colunas

```python
list(df4.columns.values)
```

```python
df4.insert(0, 'YearlyWageInDollar', df4.pop('YearlyWageInDollar'))
```

```python
list(df4.columns.values)
```

### Exportando dataframe

```python
df4.to_csv('survey_results_salary_numeric.csv', sep=',', encoding='utf-8')
```
